**Import pandas and wikipedia libaries to scrape the HTML page**

```python
import pandas as pd
import wikipedia as wp
```

**Scrape HTML page and store the table to a dataframe**

```python
#Get the html source
html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")

df = pd.read_html(html)[0]
```

```python
df.head(20)
```

**Set the correct headers for the dataframe**

```python
new_header = df.iloc[0] #grab the first row  as the header
df = df[1:] #take the current data less the header row
df.columns = new_header #set the header row as the df header
```

**Check the existing shape of the dataframe before pre-processing**

```python
df.shape
```

**Drop all the rows where Borough is not assigned**

```python
df = df.drop(df[df['Borough']=='Not assigned'].index)
```

```python
df.head(15)
```

**Reset index of the dataframe**

```python
df.reset_index(drop=True, inplace=True)
```

**Check the  current shape of the dataframe after dropping not assigned boroughs**

```python
df.shape
```

**Sort dataframe in ascending order and reset index**

```python
df= df.sort_values(by=['Postcode','Borough','Neighbourhood'], ascending=[1,1,1]).reset_index(drop=True)
df
```

**Combine rows with the same postcode but with different neighbourhood**

```python
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df
```

**For a 'Non assigned' neighbourhood, assign borough name to it**

```python
for index, row in df.iterrows():
    if row['Neighbourhood'] == 'Not assigned':
        row['Neighbourhood'] = row['Borough']
        
df
```

**Check the shape of the dataframe**

```python
df.shape
```

**Export to CSV**

```python
df.to_csv('clean_df.csv',index=False)
```
